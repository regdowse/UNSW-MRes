# Sample Eddy Matcher

Map old sample eddy IDs onto the vertically checked dataset.


```python
import pandas as pd

import seacofs_tilt_tools as tilt

paths = tilt.Paths()
df_eddies_old = pd.read_pickle(paths.old_eddies)
df_eddies, df_tilt = tilt.load_tilt_tables(paths)
sample_eddies_old = [100, 627, 1269, 2609, 34, 281, 2334, 3126]

matches = tilt.match_old_eddies(sample_eddies_old, df_eddies_old, df_eddies)
matches
```

```python
sample_eddies = matches.new_eddy.dropna().astype(int).tolist()
sample_eddies
```
