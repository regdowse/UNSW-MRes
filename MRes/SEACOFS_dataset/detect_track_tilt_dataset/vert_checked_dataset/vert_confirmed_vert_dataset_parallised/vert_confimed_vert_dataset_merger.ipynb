```python
import numpy as np
import pandas as pd
import os
import pickle

def merge_dic_vert(*dics):
    out = {}

    for dic in dics:
        for eddy_key, day_dic in dic.items():
            out.setdefault(eddy_key, {})

            for day_key, df in day_dic.items():
                if day_key in out[eddy_key]:
                    print(f'Warning: overwriting {eddy_key} {day_key}')

                out[eddy_key][day_key] = df

    return out

base_dir = '/srv/scratch/z5297792/SEACOFS_26yr_eddy_dataset/DOPPIO_SEACOFS_26yr_50m_vert_check/dic_vert_parts'

part_files = [
    f'{base_dir}/dic_vert_50m_vert_check_part{worker_id}.pkl' for worker_id in range(10)
]

dics = []

for file in part_files:
    if os.path.exists(file) and os.path.getsize(file) > 0:
        with open(file, 'rb') as f:
            dics.append(pickle.load(f))
        print(f'Loaded {file}')
    else:
        print(f'Skipping missing/empty file: {file}')

dic = merge_dic_vert(*dics)

out_file = '/srv/scratch/z5297792/SEACOFS_26yr_eddy_dataset/DOPPIO_SEACOFS_26yr_50m_vert_check/dic_vert_doppio_50m_vert_checked.pkl'

tmp_file = out_file #+ '.tmp'
with open(tmp_file, 'wb') as f:
    pickle.dump(dic, f)

os.replace(tmp_file, out_file)

print(f'Saved merged dictionary to {out_file}')
print(f'Total eddies: {len(dic)}')
print(f'Total eddy-days: {sum(len(v) for v in dic.values())}')
```
