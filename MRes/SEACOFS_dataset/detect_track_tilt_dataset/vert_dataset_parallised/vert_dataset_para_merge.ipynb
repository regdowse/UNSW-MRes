```python
def merge_dic_vert(*dics):
    out = {}

    for dic in dics:
        for eddy_key, day_dic in dic.items():
            out.setdefault(eddy_key, {})

            for day_key, df in day_dic.items():
                if day_key in out[eddy_key]:
                    print(f'Warning: overwriting {eddy_key} {day_key}')

                out[eddy_key][day_key] = df

    return out
    
```

```python
with open(part1_file, 'rb') as f:
    dic1 = pickle.load(f)

with open(part2_file, 'rb') as f:
    dic2 = pickle.load(f)

dic = merge_dic_vert(dic1, dic2)

with open('/srv/scratch/z5297792/SEACOFS_26yr_eddy_dataset/dic_vert_doppio.pkl', 'wb') as f:
    pickle.dump(dic, f)\\
    
```
