```python
import numpy as np
import netCDF4 as nc

fname = '/srv/scratch/z3533156/26year_BRAN2020/outer_avg_01461.nc'
dataset = nc.Dataset(fname)
zeta0 = dataset.variables['zeta']
zeta0 = np.transpose(zeta0, axes=(2, 1, 0))

zeta_sum = zeta0

fnumbers = [f'{num:05}' for num in range(1491, 10611+1, 30)]

for fnumber in fnumbers:
    fname = '/srv/scratch/z3533156/26year_BRAN2020/outer_avg_'+str(fnumber)+'.nc'
    dataset = nc.Dataset(fname)
    zeta = dataset.variables['zeta']
    zeta = np.transpose(zeta, axes=(2, 1, 0))

    zeta_sum = zeta_sum + zeta

    print(fnumber)
    
    del zeta

zeta_ave = zeta_sum / (len(fnumbers)+1)

zeta_ave = np.mean(zeta_ave, axis=2)

np.save('/srv/scratch/z5297792/Climatology/zeta_ave.npy', zeta_ave)
```
