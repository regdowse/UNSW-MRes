```python
import numpy as np

fnumbers = range(1461, 10641+1, 30)

sum_all = None
cnt_all = None

for fnum in fnumbers:
    v = np.load(f'/srv/scratch/z5297792/Climatology/u_v/v_{fnum:05}.npy')
    
    if sum_all is None:
        sum_all = np.nansum(v, axis=3)
        cnt_all = np.sum(~np.isnan(v), axis=3).astype(np.int64)
    else:
        sum_all += np.nansum(v, axis=3)
        cnt_all += np.sum(~np.isnan(v), axis=3)
    print(fnum)

v_ave = np.divide(sum_all, cnt_all, out=np.full_like(sum_all, np.nan), where=cnt_all!=0)
np.save('/srv/scratch/z5297792/Climatology/v_ave.npy', v_ave)
```
